# **Problem Statement:**
Find all the list of products whose sum-of-price is between 290 and 310.

ProductList = {'p1':10, 'p2':15, 'p3':20, 'p4':25, 'p5':30, 'p6':35, 'p7':50,
               'p8':40, 'p9':55, 'p10':60, 'p11':65, 'p12':75, 'p13':70,
               'p14':45}

```python
import random as r

ProductList = {'p1':10, 'p2':15, 'p3':20, 'p4':25, 'p5':30, 'p6':35, 'p7':50,'p8':40, 'p9':55, 'p10':60, 'p11':65, 'p12':75, 'p13':70,'p14':45}
LB = 290
UB = 310
iterations = 500000
product_keys = list(ProductList.keys())
AnswerSet = set()

for i in range(iterations):
  SubSetSize = r.randint(1,len(product_keys))
  SubSet = r.sample(product_keys,SubSetSize)
  total_price = sum(ProductList[r] for r in SubSet)
  if LB <= total_price <= UB:
    AnswerSet.add(tuple(sorted(SubSet)))
  else:
    pass
AnswerList = [list(item) for item in AnswerSet]

print("All lists :")
for combo in AnswerList:
  print(combo)
print("Total Lists =", len(AnswerList))
```
